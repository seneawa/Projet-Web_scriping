Ce code suivant lit un fichier HTML, analyse son contenu pour extraire des informations spÃ©cifiques sur les villes, telles que les prix, les horaires, les lieux de dÃ©part, les numÃ©ros de tÃ©lÃ©phone, et les adresses, puis affiche ces informations de maniÃ¨re structurÃ©e.

```python
from bs4 import BeautifulSoup
import re

# Chemin du fichier HTML
html_path = r'/content/html_code.html'

# Lire le contenu du fichier HTML
with open(html_path, 'r', encoding='utf-8') as file:
    html_content = file.read()

# Analyser le contenu HTML avec BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Fonction pour extraire les numÃ©ros de tÃ©lÃ©phone
def extract_phone_numbers(text):
    phone_numbers = re.findall(r'\+?\d[\d\s\-()]{7,}\d', text)
    return phone_numbers

# Fonction pour nettoyer le texte extrait en supprimant les indicateurs
def clean_text(text, indicator):
    if indicator in text:
        return text.split(indicator, 1)[-1].strip()
    return text

# Fonction pour extraire le prix sans le texte "Prix:"
def extract_price(text):
    if 'Prix' in text:
        start = text.find('Prix:')
        text_after_price = text[start + len('Prix:'):].strip()
        price_match = re.search(r'\d+[\s\w]*FCFA', text_after_price)
        return price_match.group(0) if price_match else None
    return None

# Initialiser un dictionnaire pour stocker les donnÃ©es extraites
data = {}
seen_cities = set()  # Ensemble pour suivre les villes dÃ©jÃ  ajoutÃ©es

# Fonction pour extraire les informations de chaque section
def extract_info(section):
    heading = section.find('h2')
    if heading:
        city = heading.get_text(strip=True)
        if city in seen_cities:
            return  # Ãviter les rÃ©pÃ©titions

        seen_cities.add(city)
        info = {'Ville': city, 'Prix': None, 'Horaires et Lieu de dÃ©part': None, 'Contact': []}

        # Trouver la section de texte correspondant Ã  la ville
        text_section = section.find_next('div', class_='elementor-widget-text-editor')
        if text_section:
            text_content = text_section.get_text(separator='\n', strip=True)

            # Extraire le prix sans le texte "Prix:"
            info['Prix'] = extract_price(text_content)

            # Extraire les horaires et lieu de dÃ©part
            horaires_lieu_text = None
            if 'Horaires et' in text_content:
                start = text_content.find('Horaires et')
                end = text_content.find('Lieu de dÃ©part', start) if 'Lieu de dÃ©part' in text_content else len(text_content)
                horaires_lieu_text = text_content[start:end].strip()
                # Nettoyer le texte pour ne garder que les horaires et les lieux de dÃ©part
                cleaned_horaires_lieu_text = clean_text(horaires_lieu_text, 'Horaires et')

                # Extraire le lieu de dÃ©part
                if 'Lieu de dÃ©part' in text_content:
                    start = text_content.find('Lieu de dÃ©part')
                    lieu_de_depart_text = text_content[start:].strip()
                    end = lieu_de_depart_text.find('Adresse')
                    if end != -1:
                        lieu_de_depart_text = lieu_de_depart_text[:end].strip()
                    # Extraire et ajouter les numÃ©ros de tÃ©lÃ©phone dans le champ "Contact"
                    phone_numbers = extract_phone_numbers(lieu_de_depart_text)
                    info['Contact'].extend(phone_numbers)
                    for phone_number in phone_numbers:
                        lieu_de_depart_text = lieu_de_depart_text.replace(phone_number, '').strip()

                    # Mettre Ã  jour le champ 'Horaires et Lieu de dÃ©part' avec le lieu de dÃ©part extrait
                    if cleaned_horaires_lieu_text:
                        info['Horaires et Lieu de dÃ©part'] = f"{cleaned_horaires_lieu_text} {lieu_de_depart_text}"
                    else:
                        info['Horaires et Lieu de dÃ©part'] = lieu_de_depart_text

                # Assurer que 'Horaires et Lieu de dÃ©part' ne contient pas de texte rÃ©pÃ©titif
                if info['Horaires et Lieu de dÃ©part']:
                    info['Horaires et Lieu de dÃ©part'] = clean_text(info['Horaires et Lieu de dÃ©part'], 'Lieu de dÃ©part')

            # Extraire l'adresse (si prÃ©sent)
            if 'Adresse' in text_content:
                start = text_content.find('Adresse')
                address_text = text_content[start:].strip()
                info['Adresse'] = clean_text(address_text, 'Adresse:')

            # Convertir la liste de contacts en une chaÃ®ne de caractÃ¨res
            info['Contact'] = ', '.join(info['Contact'])

        return info

# Extraire les informations pour chaque ville
for section in soup.find_all('div', class_='elementor-element'):
    city_info = extract_info(section)
    if city_info:
        data[city_info['Ville']] = city_info

# Afficher les donnÃ©es extraites
for city, info in data.items():
    print(f"Ville: {city}")
    for key, value in info.items():
        if key != 'Ville':  # Exclure le titre de la ville dans l'affichage des dÃ©tails
            print(f"  {key}: {value}")
    print()
```
